```python
df['BMI Category'] =df['BMI Category'].replace({'Normal Weight':'Normal'})
df['BMI Category'].value_counts()
```

```python
# Split 'Blood Pressure' into 'Systolic' and 'Diastolic'
df['Systolic'] = df['Blood Pressure'].apply(lambda x: int(x.split('/')[0]))
df['Diastolic'] = df['Blood Pressure'].apply(lambda x: int(x.split('/')[1]))

# Define function to categorize blood pressure
def bloodpressure_category(row):
    if row['Systolic'] < 120 and row['Diastolic'] < 80:
        return 'Normal'
    elif 120 <= row['Systolic'] < 130 and row['Diastolic'] < 80:
        return 'Elevated'
    else:
        return 'Hypertension'

# Apply the function to create a new category
df['Blood Pressure Category'] = df.apply(bloodpressure_category, axis=1)

# Drop the temporary columns
df.drop(columns=['Systolic', 'Diastolic','Blood Pressure'], inplace=True)

# Display a random sample of the updated dataframe
df.sample()
```

```python
columns_to_check = ['Gender', 'Occupation', 'Quality of Sleep',
        'Stress Level', 'BMI Category','Blood Pressure Category']
columns_per_row = 3

# Calculate total rows needed
total_rows = (len(columns_to_check) + columns_per_row - 1) // columns_per_row

# Create subplots
fig, axes = plt.subplots(total_rows, columns_per_row, figsize=(18, 6 * total_rows))  # Adjust figure size as needed
axes = axes.flatten()  # Flatten axes for easy indexing

for idx, column in enumerate(columns_to_check):
    # Grouped data
    grouped_data = df.groupby(column)['Sleep Disorder'].value_counts(normalize=True).unstack()

    # Plot
    grouped_data.plot(kind='bar', stacked=True, ax=axes[idx], legend=(idx == 0))
    axes[idx].set_title(f'{column} vs Sleep Disorder Proportions')
    axes[idx].set_xlabel(column)
    axes[idx].set_ylabel('Proportion')

    # Add legend only for the first plot
    if idx == 0:
        axes[idx].legend(title='Sleep Disorder')
    else:
        axes[idx].legend().remove()

# Hide unused subplots
for ax in axes[len(columns_to_check):]:
    ax.axis('off')

plt.tight_layout()
plt.show()

```

```python
range_steps = {
    'Age': 10,
    'Sleep Duration': 1,
    'Physical Activity Level': 5,
    'Heart Rate': 5,
    'Daily Steps': 1000
}

df_transformed = df.copy()

def value_to_range(value, step):
    start = (value // step) * step
    end = start + step
    return f"{start}-{end}"

for col, step in range_steps.items():
    if col in df_transformed.columns:
        df_transformed[f'{col}_Range'] = df_transformed[col].apply(value_to_range, step=step)

# List of columns to check for plotting
columns_to_check = ['Age', 'Sleep Duration', 'Physical Activity Level', 'Heart Rate', 'Daily Steps']
columns_per_row = 3

# Calculate total rows needed for subplots
total_rows = (len(columns_to_check) + columns_per_row - 1) // columns_per_row

# Create subplots
fig, axes = plt.subplots(total_rows, columns_per_row, figsize=(18, 6 * total_rows))
axes = axes.flatten()  # Flatten axes for easy indexing

# Loop through columns to plot the data
for idx, column in enumerate(columns_to_check):
    # Grouped data using the range columns
    grouped_data = df_transformed.groupby(f'{column}_Range')['Sleep Disorder'].value_counts(normalize=True).unstack()

    # Plot the data
    grouped_data.plot(kind='bar', stacked=True, ax=axes[idx], legend=(idx == 0))
    axes[idx].set_title(f'{column} vs Sleep Disorder Proportions')
    axes[idx].set_xlabel(f'{column} Ranges')
    axes[idx].set_ylabel('Proportion')

    if idx == 0:
        axes[idx].legend(title='Sleep Disorder')
    else:
        axes[idx].legend().remove()

for ax in axes[len(columns_to_check):]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
disorder_counts = (
    df.groupby(['Sleep Disorder', 'Occupation'])
    .size()
    .reset_index(name='Count')
)

unique_disorders = df['Sleep Disorder'].unique()[:3]

fig, axs = plt.subplots(1, 3, figsize=(18, 6))

for i, disorder in enumerate(unique_disorders):

    disorder_data = disorder_counts[disorder_counts['Sleep Disorder'] == disorder]

    disorder_data['Percentage'] = (
        disorder_data['Count'] / disorder_data['Count'].sum() * 100
    )

    axs[i].pie(
        disorder_data['Percentage'],
        labels=disorder_data['Occupation'],
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Paired.colors
    )
    axs[i].set_title(f'{disorder} by Occupation')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X=df.drop(columns=['Sleep Disorder'])
y =df['Sleep Disorder']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

Lr = LogisticRegression()

Lr.fit(X_train,y_train)

y_pred = Lr.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test, y_pred)
print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(cm)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

Rf = RandomForestClassifier()

Rf.fit(X_train,y_train)

y_pred = Rf.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test, y_pred)
print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(cm)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

svc = SVC()

svc.fit(X_train,y_train)

y_pred = svc.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test, y_pred)
print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(cm)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = [
    {'n_estimators' : [100, 200, 300, 400],'criterion' : ['gini', 'entropy']}
]

grid = GridSearchCV(Rf, param_grid, cv=5, scoring='accuracy', return_train_score=True)
grid.fit(X_train, y_train.values.ravel())
```
